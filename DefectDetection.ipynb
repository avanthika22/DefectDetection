````python
import os
os.environ["OPENAI_API_KEY"] = "your-openai-api-key"
````

````python
from openai import OpenAI
import json


def analyze_defect(defect_type: str):
    """
    Uses GPT to analyze a given defect type and return cause & fix.

    """
    from openai import OpenAI
    import json, re

    client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))

    system_prompt = (
        "You are an industrial defect analysis assistant. "
        "The user will provide a defect type (like crack, corrosion, dent, discoloration, etc.). "
        "You must return a JSON object with: likely_cause, recommended_action, and confidence (0-1)."
    )

    user_prompt = f"Defect type: {defect_type}\nRespond only as JSON."

    response = client.chat.completions.create(
        model="gpt-4o-mini",  # or gpt-3.5-turbo if preferred
        messages=[
            {"role": "system", "content": system_prompt},
            {"role": "user", "content": user_prompt}
        ],
        temperature=0.2
    )

    text = response.choices[0].message.content.strip()

    # ðŸ§¹ Clean markdown code fences like ```json ... ```
    cleaned = re.sub(r"```(?:json)?", "", text).replace("```", "").strip()

    try:
        return json.loads(cleaned)
    except Exception:
        return {"error": "Could not parse response", "raw_output": cleaned}

````

````python
print("ðŸ”§ Welcome to the AI Defect Advisor!")
print("Type 'exit' to quit.\n")

while True:
    defect_type = input("Enter a defect type (e.g., crack, corrosion, dent, discoloration): ").strip()
    if defect_type.lower() in ["exit", "quit"]:
        print("\nðŸ‘‹ Exiting Defect Advisor. Goodbye!")
        break

    print("\nAnalyzing defect, please wait...\n")
    result = analyze_defect(defect_type)

    print("ðŸ“‹ Analysis Result:")
    print(json.dumps(result, indent=2))
    print("\n-----------------------------------\n")
````
